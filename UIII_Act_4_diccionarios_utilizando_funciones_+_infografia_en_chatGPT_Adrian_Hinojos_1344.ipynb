#### Cajero

#### Atributos:

* Saldo inicial
* Ventas procesadas
* Tienda
* Arqueo de caja
* Tiempo de servicio

```python
numero = 123
saldo = 1000.00
ventas = 50
tienda = "SuperMart"
arqueo = "Pendiente"
tiempo = "8 horas"

# Diccionario del cajero
cajero = {
    "numeroCaja": numero,
    "SaldoInicial": saldo,
    "VentasProcesadas": ventas,
    "Tienda": tienda,
    "ArqueoCaja": arqueo,
    "TiempoServicio": tiempo
}
```

```python
# Acceder usando llaves []
cajero["numeroCaja"]
cajero["SaldoInicial"]
cajero["VentasProcesadas"]
cajero["Tienda"]
cajero["ArqueoCaja"]
cajero["TiempoServicio"]
```

```python
# Actualizar usando llaves []
cajero["SaldoInicial"] = input("Nuevo saldo inicial: ")
cajero["VentasProcesadas"] = input("Nuevas ventas procesadas: ")
cajero["Tienda"] = input("Nueva tienda: ")
cajero["ArqueoCaja"] = input("Nuevo arqueo: ")
cajero["TiempoServicio"] = input("Nuevo tiempo de servicio: ")
```
