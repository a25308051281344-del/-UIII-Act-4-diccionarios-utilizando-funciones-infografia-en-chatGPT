```python
# ============================================
# Unidad III Act3
# Tema: Cajero
# Alumno: Adrian Hinojos Chavez 1344
# ============================================

# Lista vacía
diccionario_cajero = ()

# --------------------------------------------
# Función agregar cajero
# --------------------------------------------
def agregar_cajero():

    numero = input("Número de caja: ")
    saldo = input("Saldo inicial: ")
    ventas = input("Ventas procesadas: ")
    tienda = input("Tienda: ")
    arqueo = input("Arqueo de caja: ")
    tiempo = input("Tiempo de servicio: ")

    cajero = {
        "NumeroCaja": numero,
        "SaldoInicial": saldo,
        "VentasProcesadas": ventas,
        "Tienda": tienda,
        "ArqueoCaja": arqueo,
        "TiempoServicio": tiempo
    }

    diccionario_cajero.append(cajero)

    print("\nCAJERO AGREGADO\n")

    listar_cajero()


# --------------------------------------------
# Función listar cajeros
# --------------------------------------------
def listar_cajero():

    print("\n================ LISTA DE CAJEROS ================\n")

    print("{:<15} {:<15} {:<20} {:<15} {:<15} {:<20}".format(
        "Caja",
        "Saldo",
        "Ventas",
        "Tienda",
        "Arqueo",
        "Tiempo"
    ))

    print("-" * 100)

    for cajero in diccionario_cajero:

        print("{:<15} {:<15} {:<20} {:<15} {:<15} {:<20}".format(
            cajero["NumeroCaja"],
            cajero["SaldoInicial"],
            cajero["VentasProcesadas"],
            cajero["Tienda"],
            cajero["ArqueoCaja"],
            cajero["TiempoServicio"]
        ))


# --------------------------------------------
# Función buscar cajero
# --------------------------------------------
def buscar_cajero():

    buscar = input("Ingrese número de caja a buscar: ")

    print("\n============== RESULTADO DE BÚSQUEDA ==============\n")

    for cajero in diccionario_cajero:

        if cajero["NumeroCaja"] == buscar:

            for key, value in cajero.items():
                print(key, ":", value)

    listar_cajero()


# --------------------------------------------
# Función borrar cajero
# --------------------------------------------
def borrar_cajero():

    borrar = input("Ingrese número de caja a borrar: ")

    for cajero in diccionario_cajero:

        if cajero["NumeroCaja"] == borrar:

            diccionario_cajero.remove(cajero)

            print("\nCAJERO ELIMINADO\n")

    listar_cajero()


# --------------------------------------------
# Función actualizar cajero
# --------------------------------------------
def actualizar_cajero():

    actualizar = input("Ingrese número de caja a actualizar: ")

    for cajero in diccionario_cajero:

        if cajero["NumeroCaja"] == actualizar:

            cajero["SaldoInicial"] = input("Nuevo saldo inicial: ")
            cajero["VentasProcesadas"] = input("Nuevas ventas procesadas: ")
            cajero["Tienda"] = input("Nueva tienda: ")
            cajero["ArqueoCaja"] = input("Nuevo arqueo: ")
            cajero["TiempoServicio"] = input("Nuevo tiempo de servicio: ")

            print("\nCAJERO ACTUALIZADO\n")

    listar_cajero()


# ============================================
# MENÚ PRINCIPAL
# ============================================

while True:

    print("\n=========== MENÚ CAJERO ===========")
    print("1 Agregar cajero")
    print("2 Listar cajeros")
    print("3 Buscar cajero")
    print("4 Borrar cajero")
    print("5 Actualizar cajero")
    print("6 Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_cajero()

    elif opcion == "2":
        listar_cajero()

    elif opcion == "3":
        buscar_cajero()

    elif opcion == "4":
        borrar_cajero()

    elif opcion == "5":
        actualizar_cajero()

    elif opcion == "6":
        print("\nPrograma finalizado")
        break

    else:
        print("\nOpción incorrecta")
```
